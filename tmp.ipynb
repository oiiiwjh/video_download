## interval 删除 sample后的结果


```python
import numpy as np

def get_rubbish_csv(csv_path, interval=5, min_num=150):
	df = pd.read_csv(csv_path)
	df['base_id'] = df['id'].apply(lambda x: '_'.join(x.split('_')[:-1]))
	df['n'] = df['id'].apply(lambda x: int(x.split('_')[-1]))
	df = df.sort_values(by=['base_id', 'n'], ascending=[True, True])
	result = []
	for _, group in df.groupby('base_id'):
		# print(len(group))
		if len(group) < min_num:
			# 将小于min_num的组全部保留
			result.append(group)
			print(f'group {group.iloc[0]["base_id"]} has {len(group)} samples, less than {min_num}, keep all')
			continue
		idxs = [0]
		last_n = group.iloc[0]['n']
		for i in range(1, len(group)):
			if group.iloc[i]['n'] >= last_n + interval:
				idxs.append(i)
				last_n = group.iloc[i]['n']
		result.append(group.iloc[idxs])
	return pd.concat(result, ignore_index=True)
df = get_rubbish_csv('/share/wjh/megasam_sample/stage2/stage2_total_done_sample_supp_3.csv')
# df.to_csv('stage2_total_done_sample_supp_3_rubbish.csv', index=False)
save_csv = False
if save_csv:
	df.to_csv('stage2_total_done_sample_supp_3_rubbish.csv', index=False)
else:
	# only save ids
	df['id'].to_csv('stage2_total_done_sample_supp_3_rubbish_ids.csv', index=False, header=False)
```

```python
txt = '/home/wjh/step2_megasam/1.txt'
# txt 每行为step0, v1: tensor([ 0.0014, -0.0013, -0.0371], device='cuda:0'), v2: tensor([ 0.0039, -0.0030, -0.0396], device='cuda:0'), angle: 0.07228053361177444
# 得到angle的值
with open(txt, 'r') as f:
	lines = f.readlines()
	angles = []
	for line in lines:
		if 'angle' in line:
			angle = float(line.split('angle: ')[-1])
			angles.append(angle)
		else:
			continue
	print(len(angles), angles)
angles = angles[:-1]
```

```python
import matplotlib.pyplot as plt
import numpy as np
# plot angles
plt.plot(angles)
plt.xlabel('sample')
plt.ylabel('angle')
```

```python
# 对angle进行傅里叶变换
angles = np.array(angles)
angles = angles - np.mean(angles)
angles = angles / np.std(angles)
angles = np.fft.fft(angles)
angles = np.abs(angles)
plt.plot(angles)
# 绘制横纵坐标
plt.xlabel('frequency')
plt.ylabel('amplitude')
```
